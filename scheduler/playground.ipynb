```python
from scheduler.scripts import run_algo_impl

NUM_EXP = 5
sev_data = []

for i in range(NUM_EXP):
    sev_data.append(run_algo_impl("MBrainLike", 3))
```

```python
import matplotlib.pyplot as plt
import numpy as np

for data in sev_data:
    x = np.arange(0, len(data), 1)
    y1 = [elem[0] for elem in data]
    y2 = [elem[1] for elem in data]

    fig, axs = plt.subplots(1, 2, figsize=(12, 5))
    axs[0].plot(x, y1)
    axs[1].plot(x, y2)
    fig.suptitle('Title')
```
